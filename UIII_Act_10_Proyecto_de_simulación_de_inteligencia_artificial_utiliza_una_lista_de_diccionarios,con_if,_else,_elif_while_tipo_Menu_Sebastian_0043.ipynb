### Proyecto de simulación de inteligencia artificial  utiliza una lista de diccionarios,con if, else, elif  while tipo Menu
### 18. Administrador Los administradores

```python
# --- INICIALIZACIÓN ---
# La lista principal de registros empieza vacía
administrador = []
# El diccionario de trabajo para capturar datos también empieza vacío
losadministradores = {}

print("================================================")
print("   NÚCLEO IA 'ADMIN-X': GESTIÓN DE SEGURIDAD    ")
print("================================================")

while True:
    # --- LÓGICA DE PERCEPCIÓN DE LA IA ---
    # La IA usa la función len() para evaluar el estado de la lista [1]
    cantidad = len(administrador)

    # Estructura if-elif-else para determinar el estado del sistema [2, 3]
    if cantidad == 0:
        estado_ia = "ALERTA: Sin administradores. El sistema está vulnerable."
    elif cantidad < 3:
        estado_ia = "NORMAL: Equipo de administración en nivel básico."
    elif cantidad < 6:
        estado_ia = "OPTIMO: Carga de supervisión ideal."
    else:
        estado_ia = "ADVERTENCIA: Saturación de perfiles detectada."

    # --- MENÚ INTERACTIVO ---
    print(f"\n[Reporte de IA]: {estado_ia}")
    print(f"Total en lista 'administrador': {cantidad}")
    print("1. Registrar Administrador (Crear Diccionario)")
    print("2. Mostrar Base de Datos (Ciclo For)")
    print("3. Buscar Perfil Específico")
    print("4. Formatear Lista (Borrar todo)")
    print("5. Desconectar IA")

    opcion = input("\nIngrese comando numérico: ")

    if opcion == "1":
        # Creamos un diccionario vacío para los nuevos datos [4]
        losadministradores = {}

        # Entrada de datos por teclado [5, 6]
        id_admin = input("ID de empleado: ")
        nombre = input("Nombre completo: ")
        rango = input("Rango de acceso (A/B/C): ")

        # Asignamos valores al diccionario usando claves [4, 7]
        losadministradores["id"] = id_admin
        losadministradores["nombre"] = nombre
        losadministradores["rango"] = rango

        # Agregamos el diccionario 'losadministradores' a la lista 'administrador' [8]
        administrador.append(losadministradores)
        print(f">>> IA: Perfil de {nombre} guardado exitosamente.")

    elif opcion == "2":
        # Verificamos si la lista tiene elementos antes de recorrerla [9]
        if cantidad > 0:
            print("\n--- REGISTROS EN LA LISTA 'ADMINISTRADOR' ---")
            # Ciclo for para recorrer la lista de diccionarios [10-12]
            for item in administrador:
                # Accedemos a los datos internos del diccionario [7]
                print(f"ID: {item['id']} | Admin: {item['nombre']} | Nivel: {item['rango']}")
        else:
            print(">>> IA: La lista está vacía actualmente.")

    elif opcion == "3":
        busqueda = input("Ingrese el nombre del administrador a localizar: ")
        encontrado = False

        # For para buscar dentro de la lista de diccionarios [13]
        for item in administrador:
            # Comparamos ignorando mayúsculas/minúsculas [13]
            if item["nombre"].lower() == busqueda.lower():
                print(f">>> IA: ¡Localizado! El ID es {item['id']} y su rango es {item['rango']}.")
                encontrado = True

        if encontrado == False:
            print(">>> IA: No se encontraron coincidencias en la memoria.")

    elif opcion == "4":
        # Reiniciamos la lista a vacía [8]
        administrador = []
        print(">>> IA: Se ha vaciado la lista 'administrador'.")

    elif opcion == "5":
        print(">>> IA: Finalizando procesos de administración... ¡Adiós!")
        break # Rompe el ciclo while True para terminar el programa [14]

    else:
        print(">>> IA: ERROR. Comando no identificado.")

print("\n--- NÚCLEO IA FUERA DE LÍNEA ---")
```
