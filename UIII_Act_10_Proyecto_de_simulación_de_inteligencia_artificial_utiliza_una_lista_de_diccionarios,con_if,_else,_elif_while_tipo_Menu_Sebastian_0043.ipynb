### Proyecto de simulación de inteligencia artificial  utiliza una lista de diccionarios,con if, else, elif  while tipo Menu
### 18. Administrador Los administradores

```python
# --- INICIALIZACIÓN ---
# La lista principal de diccionarios empieza vacía
administrador = {}
losadministradores = []
# El diccionario de trabajo se usará para registrar datos
sistema_activo = True

print("................................................")
print("   NÚCLEO DE INTELIGENCIA ARTIFICIAL 'ADMIN-Z'  ")
print("................................................")

while True:
    # --- LÓGICA DE PERCEPCIÓN DE LA IA ---
    # La IA analiza la lista para dar un reporte de estado
    cantidad = len(losadministradores)

    if cantidad == 0:
        estado_ia = "ALERTA: Servidor sin vigilancia humana."
    elif cantidad == 1:
        estado_ia = "ESTADO: Vigilancia mínima detectada."
    elif cantidad < 4:
        estado_ia = "ESTADO: Seguridad óptima del servidor."
    else:
        estado_ia = "PRECAUCIÓN: Exceso de administradores (Saturación)."

    # --- MENÚ INTERACTIVO ---
    print(f"\n[IA Reporta]: {estado_ia}")
    print(f"Total Administradores: {cantidad}")
    print("1. Registrar Administrador (Nuevo Diccionario)")
    print("2. Listar Equipo (Ciclo For)")
    print("3. Buscar Administrador")
    print("4. Reiniciar Base de Datos")
    print("5. Apagar Núcleo IA")

    opcion = input("\nInstrucción para la IA: ")

    if opcion == "1":
        # Creación de un diccionario vacío para los datos del nuevo administrador
        administrador = {}
        nombre = input("Ingrese nombre: ")
        nivel = input("Ingrese nivel de acceso (1-10): ")

        # Guardamos los datos en el diccionario utilizando claves
        administrador["nombre"] = nombre
        administrador["acceso"] = nivel

        # Agregamos el diccionario a la lista principal
        losadministradores.append(administrador)
        print(f">>> IA: Administrador {nombre} integrado al núcleo.")

    elif opcion == "2":
        if cantidad > 0:
            print("\n--- LISTA DE ADMINISTRADORES REGISTRADOS ---")
            # Uso de CICLO FOR para recorrer la lista de diccionarios
            for admin in losadministradores:
                print(f"Nombre: {admin['nombre']} | Acceso: {admin['acceso']}")
        else:
            print(">>> IA: La lista está vacía. No hay datos para mostrar.")

    elif opcion == "3":
        busqueda = input("Nombre del administrador a buscar: ")
        encontrado = False

        # Usamos for para buscar dentro de los diccionarios de la lista
        for admin in losadministradores:
            if admin["nombre"].lower() == busqueda.lower():
                print(f">>> IA: Localizado. Su nivel de acceso es: {admin['acceso']}")
                encontrado = True

        if encontrado == False:
            print(">>> IA: Usuario no encontrado en los registros.")

    elif opcion == "4":
        # Limpiamos la lista para que vuelva a estar vacía
        losadministradores = []
        print(">>> IA: Memoria purgada. Lista de diccionarios vacía.")

    elif opcion == "5":
        print(">>> IA: Desconectando protocolos... Adiós.")
        break # Rompe el ciclo While True para salir

    else:
        print(">>> IA: Comando inválido. Reintente.")

print("\n--- NÚCLEO IA FUERA DE LÍNEA ---")
```
